```python
# We will be working with the IRIS dataset (part of sample datasets in scikitlearn)
# We will classify flowers types with different classifiers 
# Here it would be LDAClassifier

from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
# d_names = iris.data_names
y_names = iris.target_names

# print(X[40:60])
# print(y[40:60])
# print(y_names)
# print(d_names)
```

```python
from sklearn.model_selection import train_test_split

# use train_test_split to split the data randomly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

```

```python
# LDA Classifier applied to classify the flowers 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

LDACL = LinearDiscriminantAnalysis()
LDACL.fit(X_train, y_train)
```

```python
# estimate accuracy for Training and Testing

print("Training Score: {}".format(LDACL.score(X_train, y_train)))
print("Testing Score: {}".format(LDACL.score(X_test, y_test)))
```

```python
# plot the confusion matrix

import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

y_pred = LDACL.predict(X_test)

fig, ax = plt.subplots(figsize=(10, 5))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, ax=ax)
ax.xaxis.set_ticklabels(y_names)
ax.yaxis.set_ticklabels(y_names)
_ = ax.set_title(
    f"Confusion Matrix for {LDACL.__class__.__name__} IRIS Classification"
)
```

```python

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('Original Features Data (Training Set)')
plt.show()
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test)
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('Original Features Data (Testing Set)')
plt.show()
```

```python
X_train_transformed = LDACL.transform(X_train)
print(X_train_transformed.shape)
plt.scatter(X_train_transformed[:, 0], X_train_transformed[:, 1], c=y_train)
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA Transformed Data (Training Set)')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(LDACL, X_train, y_train, cv=5)
scores

print("{} accuracy with a standard deviation of {}".format(scores.mean(), scores.std()))
```
